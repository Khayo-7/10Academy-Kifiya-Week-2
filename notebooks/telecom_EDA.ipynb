```python
percentage = 0.5
total_data = df["Total DL (Bytes)"].sum() + df["Total UL (Bytes)"].sum()
cumulative_data = df.groupby("IMSI")[["Total DL (Bytes)", "Total UL (Bytes)"]].sum().reset_index()
cumulative_data["Total Data (Bytes)"] = cumulative_data["Total DL (Bytes)"] + cumulative_data["Total UL (Bytes)"]
cumulative_data = cumulative_data.sort_values(by="Total Data (Bytes)", ascending=False)
cumulative_data["Cumulative Total Data"] = cumulative_data["Total Data (Bytes)"].cumsum()
users_contribution_percentage = cumulative_data[cumulative_data["Cumulative Total Data"] <= percentage * total_data]
percent_of_users = users_contribution_percentage["IMSI"].nunique() / df["IMSI"].nunique()
print(f"{percentage * 100}% of the data is generated by {percent_of_users*100:.2f}% of the users.")
```
